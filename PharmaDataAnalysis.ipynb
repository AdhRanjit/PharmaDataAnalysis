```python
#Importing  matplotlib and seaborn
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Group by 'Year' and Calculate the sum of sales
yearly_Sales = df.groupby('Year')['Sales'].sum().reset_index()

plt.bar(yearly_Sales['Year'], yearly_Sales['Sales'], color='skyblue')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Total Sales')

plt.title('Total Sales by Year')

#Labelling y-axis labels as integers
plt.ticklabel_format(style='plain', axis='y')
```

```python
# Line Plot Directly Using Matplotlib

#choosing the columns for X and Y
x_column = 'Year'
y_column = 'Sales'

#Group by year and sum of the sales
sales_by_year = df.groupby(x_column)[y_column].sum().reset_index()

#Plotting the line plot

plt.plot(sales_by_year[x_column], sales_by_year[y_column], marker ='o', linestyle='-', color='b')
plt.title("Line Plot of Total Sales Over Years")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.grid(True)

#Labelling y-axis labels as integers
plt.ticklabel_format(style='plain', axis='y')

plt.show()
```

```python
yearly_sum = df.groupby(df['Year'])['Sales'].sum()
yearly_sum.plot(kind="pie")
plt.title("Sales by Year (2017-2024)")
plt.grid(True)
```

```python
from matplotlib.ticker import FuncFormatter
# Convert 'Month' to categorical with a proper order
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
df['Month'] = pd.Categorical(df['Month'], categories=month_order, ordered=True)

# Choose the columns for X and Y axes
x_column = 'Month'
y_column = 'Sales'

# Group by month and sum the sales
sales_by_month = df.groupby(x_column)[y_column].sum().reset_index()


# Plotting the line plot with skipped month labels
plt.figure(figsize=(10, 6))
plt.plot(sales_by_month[x_column], sales_by_month[y_column], marker='o', linestyle='-', color='b')
plt.title('Line Plot of Total Sales Over Months')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.legend(["Sales"])


#Labelling y-axis labels as integers
plt.ticklabel_format(style='plain', axis='y')

# Skip every other month for better clarity
plt.xticks(rotation=45)
plt.xticks(range(0, len(month_order)), month_order)


plt.grid(True)
plt.show()
```

```python
# Convert 'Month' to categorical with a proper order
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
df['Month'] = pd.Categorical(df['Month'], categories=month_order, ordered=True)

# Choose the columns for X and Y axes
x_column = 'Sales'
y_column = 'Month'


# Function to format x-axis labels
def format_ticks(value, _):
    return '{:,.0f}'.format(value)

# Plotting the horizontal bar plot
plt.figure(figsize=(10, 6))
plt.barh(sales_by_month[y_column], sales_by_month[x_column], color='b')
plt.title('Total Sales Over Months')
plt.xlabel('Total Sales')
plt.ylabel('Month')


#Labelling y-axis labels as integers
plt.ticklabel_format(style='plain', axis='x')

plt.grid(True)
plt.show()
```

```python
# Group by 'Country' and 'Year', and calculate the total sales
total_sales_by_country = df.groupby(['Country', 'Year'])['Sales'].sum().reset_index()

# Create bar graph for each country
fig, ax = plt.subplots(figsize=(10, 6))

for country in total_sales_by_country['Country'].unique():
    country_data = total_sales_by_country[total_sales_by_country['Country'] == country]
    ax.bar(country_data['Year'], country_data['Sales'], label=country)

ax.legend()
ax.set_xlabel('Year')
ax.set_ylabel('Total Sales')
ax.grid(True)

plt.show()
```

```python
# Group by 'Country' and 'Year', and calculate the total sales
total_sales_by_country = df.groupby(['Country', 'Year'])['Sales'].sum().reset_index()

# Create line graph for each country
fig, ax = plt.subplots(figsize=(10, 6))

for country in total_sales_by_country['Country'].unique():
    country_data = total_sales_by_country[total_sales_by_country['Country'] == country]
    ax.plot(country_data['Year'], country_data['Sales'], label=country, marker='o')

ax.legend()
ax.set_xlabel('Year')
ax.set_ylabel('Total Sales')
ax.grid(True)

plt.show()
```

```python
# Group by 'Country' and calculate the total sales
total_sales_by_country = df.groupby('Country')['Sales'].sum()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(total_sales_by_country, labels=total_sales_by_country.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Set3.colors)

# Add title
plt.title('Total Sales Distribution by Country')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group by 'Country' and 'Year', and calculate the total sales
total_sales_by_country_yearly = df.groupby(['Country', 'Year'])['Sales'].sum().reset_index()

# Create subplots for each year with increased vertical spacing
unique_years = total_sales_by_country_yearly['Year'].unique()

fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(15, 8))
fig.suptitle('Total Sales Distribution by Country (2017-2024)')

# Flatten the 2D array of subplots for easier iteration
axes = axes.flatten()

for i, year in enumerate(unique_years):
    year_data = total_sales_by_country_yearly[total_sales_by_country_yearly['Year'] == year]
    
    # Create a pie chart for each year
    axes[i].pie(year_data['Sales'], labels=year_data['Country'], autopct='%1.1f%%', startangle=90, colors=plt.cm.Set3.colors)
    
    # Add title
    axes[i].set_title(f'Year {year}')

# Adjust layout for better spacing
plt.subplots_adjust(hspace=0.5, wspace=0.3)

# Show the plot
plt.show()
```

```python
# Group by 'Manager' and Calculate the sum of sales
manager_sales = df.groupby('Manager')['Sales'].sum()

# Create a bar graph
manager_sales.plot(kind='bar', color='skyblue')

# Add labels and title
plt.xlabel('Manager')
plt.ylabel('Total Sales')
plt.title('Total Sales by Manager')

plt.ticklabel_format(style='plain', axis='y')


# Show the plot
plt.show()
```

```python
# Group by 'Year' and 'Manager' and Calculate the sum of sales
yearly_manager_sales = df.groupby(['Year', 'Manager'])['Sales'].sum().unstack()

# Create subplots for each year
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 8), sharey=True)
fig.suptitle('Total Sales by Manager (2017-2024)')

# Flatten the 2D array of subplots for easier iteration
axes = axes.flatten()

# Plot a bar graph for each year
for i, (year, data) in enumerate(yearly_manager_sales.items()):
    data.plot(kind='bar', ax=axes[i], color='skyblue')
    axes[i].set_title(f'Year {year}')
    axes[i].set_xlabel('Manager')
    axes[i].set_ylabel('Total Sales')

# Adjust layout for better spacing
plt.tight_layout(rect=[0, 0, 1, 0.96])

plt.ticklabel_format(style='plain', axis='y')

# Show the plot
plt.show()
```
